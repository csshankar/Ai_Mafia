```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
##reading training data

dfx = pd.read_csv('./linear_regression/linearX.csv')
dfy = pd.read_csv('./linear_regression/linearY.csv')

dfx = dfx.values
dfy = dfy.values

x= dfx.reshape((-1,1))
y= dfy.reshape((-1,1))
print(x.shape, y.shape)
print(y)
```

```python
plt.scatter(x,y)
X = (x-x.mean())/x.std()
Y = y
plt.scatter(X,Y)
plt.show()
```

```python
print(x)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
# Training

model.fit(x,y)
```

```python
# Prediction
output = model.predict(x)
```

```python
bias = model.intercept_
coeff = model.coef_

print(bias)
print(coeff)
```

```python
model.score(x,y)
```

```python
# Visualise

plt.scatter(x,y,label = 'data')
plt.plot(x,output,color = 'black', label = 'prediciton')
plt.legend()
plt.show()
```
